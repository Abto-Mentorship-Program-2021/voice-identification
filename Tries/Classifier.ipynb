```python
from pydub import AudioSegment
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
```

```python
filepath = 'speaker0.wav'

y, sr = librosa.load(filepath)

# Normalize time series between -1 and 1
y_ns = ((y - np.min(y)) * 2) / (np.max(y) - np.min(y)) - 1
```

```python

org_len = len(y_ns)
intervals = librosa.effects.split(y_ns, top_db= 15, ref= np.max)
intervals = intervals.tolist()
y_ns = (y_ns.flatten()).tolist()
nonsilent_y = []

for p,q in intervals :
    nonsilent_y = nonsilent_y + y_ns[p:q+1] 

y_ns = np.array(nonsilent_y)

```

```python
librosa.display.waveplot(y, sr=sr, color='#515bd4')
plt.title('Before normalising and silence removing')
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
plt.show()
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
plt.show()

librosa.display.waveplot(y_ns, sr=sr, color='#515bd4')
X = librosa.stft(y_ns)
Xdb = librosa.amplitude_to_db(abs(X))
plt.show()
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
plt.show()
```

```python
X = librosa.stft(y_ns)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(12, 4))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
plt.show()

plt.figure(figsize=(12, 4))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
plt.show()


y, sr = librosa.load(filepath)
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(12, 4))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
plt.show()

plt.figure(figsize=(12, 4))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
plt.show()
```
